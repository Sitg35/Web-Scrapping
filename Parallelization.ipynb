```python
def get_url(url):
    
    r = requests.get(url)
                            
    soup = BeautifulSoup(r.text, 'html.parser')
    
    return soup
```

```python
def get_DetailPages(START_URL):
    list_of_DetailPages = []
    next_url = START_URL
    while next_url != 0:
        parsed_url = get_url(next_url)
        for name in parsed_url.find_all("div",{"style":"position:relative"}):
            inside_url = 'https://motos.coches.net' + name.select('a')[0]['href']
            list_of_DetailPages.append(inside_url)
        
        try:
            next_url = 'https://motos.coches.net/ocasion/default.aspx' + parsed_url.find("a",{"class":"pnext"}).get("href") 
        except:
             next_url = 0
    
    return list_of_DetailPages
    
```

```python
def get_color_inside(soup):
    
    color = soup.find("div",{"style":"width: 30%"}).select("li")[1].text
    
    return color
```

```python
def get_guarantee_inside(soup):
    
    guarantee = soup.find("div",{"style":"width: 30%"}).select("li")[2].text
    
    return guarantee
```

```python
def get_description_inside(soup):
    
    description = soup.find("div",{"id":"div_comentarios"}).select("p")
    
    return description
```

```python
def get_title_inside(soup):
    
    title = soup.find("span",{"class":"floatleft"}).text
    
    return title
```

```python
def get_price_inside(soup):
    
    price =  soup.find("p",{"class":"pvp"})
    x = price.text
       
    return x
```

```python
def get_KM_inside(soup):
    
    KM = soup.find("div",{"style":"width: 30%"}).select("li")[0].text
    
    return KM
```

```python
def get_year_inside(soup):
    
    year = soup.find_all("div",{"style":"width: 33%"})[1]
    year = year.select("li")[2].text
    
    return year
```

```python
def get_location_inside(soup):
    
    location = soup.find("ul",{"style": "width: 35%; margin-right: 20px"}).select("li")[3].text
    
    return location
```

```python
def genuridatas(url):
    
    data = []
    
    #parsed_url = get_url(url)
    r = requests.get(url)
                            
    parsed_url = BeautifulSoup(r.text, 'html.parser')
    
    title = get_title_inside(parsed_url)
    data.extend(title)
    price = get_price_inside(parsed_url)
    data.extend(price)
    KM = get_KM_inside(parsed_url)
    data.extend(KM)
    year = get_year_inside(parsed_url)
    data.extend(year)
    location = get_location_inside(parsed_url)
    data.extend(location)
    color = get_color_inside(parsed_url)
    data.extend(color)
    guarantee = get_guarantee_inside(parsed_url)
    data.extend(guarantee)
    description = get_description_inside(parsed_url)
    data.extend(description)
    
    
    return data
```
